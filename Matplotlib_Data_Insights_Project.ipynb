```python
import matplotlib.pyplot as plt
import numpy as np
students = ["Ayan", "Ali", "Usman", "Hamza", "Sara", "Hassan", "Areeba", "Zain"]
study_hours = [4, 6, 3, 5, 7, 2, 6, 4]
attendance = [90, 95, 80, 85, 98, 70, 92, 88]
marks = [78, 88, 65, 75, 95, 55, 90, 80]
sleep_hours = [6, 7, 5, 6, 8, 4, 7, 6]
```

```python

best_index = np.argmax(marks)
print("Top Performer Student:", students[best_index])
print("Marks:", marks[best_index])
print("Attendance:", attendance[best_index], "%")
print("Study Hours:", study_hours[best_index])
```

```python
plt.figure(figsize=(10,5))
plt.plot(students, marks, marker="o", label="Marks")
plt.plot(students, attendance, marker="s", label="Attendance")
plt.plot(students, study_hours, marker="^", label="Study Hours")
plt.title("Student Performance Dashboard (Multi-Line Plot)")
plt.xlabel("Students")
plt.ylabel("Values")
plt.grid(True)
plt.legend()
plt.show()
```

```python


high_attendance = sum(1 for x in attendance if x >= 90)
medium_attendance = sum(1 for x in attendance if 80 <= x < 90)
low_attendance = sum(1 for x in attendance if x < 80)

categories = ["High (>=90%)", "Medium (80-89%)", "Low (<80%)"]
values = [high_attendance, medium_attendance, low_attendance]

plt.figure(figsize=(7,7))
plt.pie(values, labels=categories, autopct="%1.1f%%", startangle=90)
plt.title("Attendance Category (Pie Chart)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(marks, bins=5)
plt.title("Marks Distribution (Histogram)")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(study_hours, marks, s=100)
plt.title("Study Hours vs Marks (Scatter Plot)")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(students, study_hours, marker="o")
plt.title("Study Hours Trend (Line Plot)")
plt.xlabel("Students")
plt.ylabel("Study Hours per Day")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(students, marks)
plt.title("Student Marks Report (Bar Chart)")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.show()
```
